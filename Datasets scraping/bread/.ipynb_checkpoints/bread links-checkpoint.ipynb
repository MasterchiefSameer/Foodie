```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
import pandas as pd
from IPython.display import HTML
```

```python
file_path = 'bread_1.html'
with open(file_path, 'r', encoding='utf-8') as f:
    html_content = f.read()
```

```python
soup = BeautifulSoup(html_content, 'lxml')
```

```python
t = soup.find_all('h2', class_='entry-title')
```

```python
dist = {}
for i in t:
    x = i.find('a')
    dist[x.get_text()] = x.get('href')
```

```python
dist
```

```python
df_index = pd.DataFrame.from_dict(dist, orient='index')
```

```python
df_index
```

```python
df_index.to_csv('bread_links.csv')
```
