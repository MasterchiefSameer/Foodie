```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
import pandas as pd
from IPython.display import HTML
```

```python
final_dict = {}
no_of_page = 2
for i in range(1 ,no_of_page + 1):
    
    #opening the page
    n = str(i)
    ext = '.html'
    file_path = 'daal_' + n + ext
    with open(file_path, 'r', encoding='utf-8') as f:
        html_content = f.read()
        
    #scraping the page    
    soup = BeautifulSoup(html_content, 'lxml')
    t = soup.find_all('h2', class_='entry-title')
    dict = {}
    for j in t:
        x = j.find('a')
        dict[x.get_text()] = x.get('href')
    #merging the disctt to final dict
    final_dict.update(dict)
    print(f"page " + str(i) + " done")
```

```python
df_index = pd.DataFrame.from_dict(final_dict, orient='index')
```
