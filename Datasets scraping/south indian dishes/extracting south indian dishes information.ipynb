```python
import pandas as pd
import numpy as mp
import requests
from bs4 import BeautifulSoup
```

```python
import time
```

```python
df = pd.read_csv('south_indian_dishes.csv')
```

```python
df.rename(columns={'0': 'link'}, inplace=True)
```

```python
link = df['link']
```

```python
name = []
image = []
description = []
prep_time = []
cook_time = []
total_time = []
servings_number = []
servings_unit = []
course = []
cuisine = []
diet = []
ingredients = []
instructions = []
nutrition = []
```

```python
count = 1;
for url in link:
    print(url)
    time.sleep(1)
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Referer': 'https://www.google.com/'
    }
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    a = soup.find('div', class_='wprm-recipe-template-dassanas-veg-recipes-header-text')
    try:
        name.append(a.find('h2', class_='wprm-recipe-name wprm-block-text-bold').get_text(strip=True))
    except:
        continue
    try:
        # image.append(a.find("div", class_="wprm-recipe-image").find("img").get('data-lazy-src'))
        img_tag = soup.select_one(".wprm-recipe-image img")

        if img_tag:
            img_url = (
                img_tag.get("data-lazy-src")
                or img_tag.get("data-lazy-srcset")
                or img_tag.get("srcset")
            )
            image.append(img_url)
    except:
        image.append(None)
    description.append(a.find('div', class_='wprm-recipe-summary wprm-block-text-normal').get_text(strip=True))
    soup1 = soup.find('div', class_='wprm-recipe-template-dassanas-veg-recipes-meta')
    try:
        prep_time.append(soup1.find('span', class_='wprm-recipe-prep_time').get_text(strip=True))
    except:
        prep_time.append(None)
    try:
        cook_time.append(soup1.find('span', class_='wprm-recipe-cook_time').get_text(strip=True))
    except:
        cook_time.append(None)
    try:
        total_time.append(soup1.find('span', class_='wprm-recipe-total_time').get_text(strip=True))
    except:
        total_time.append(None)
    try:
        servings_number.append(soup1.find('span', class_='wprm-recipe-servings').get_text(strip=True))
    except:
        servings_number.append(None)
    try:
        servings_unit.append(soup1.find('span', class_='wprm-recipe-servings-unit').get_text(strip=True))
    except:
        servings_unit.append(None)
    try:
        course.append(soup1.find('span', class_='wprm-recipe-course').get_text(strip=True))
    except:
        course.append(None)
    try:
        cuisine.append(soup1.find('span', class_='wprm-recipe-cuisine').get_text(strip=True))
    except:
        cuisine.append(None)
    try:
        diet.append(soup1.find('span', class_='wprm-recipe-suitablefordiet').get_text(strip=True))
    except:
        diet.append(None)
    try:
        ingredients.append(soup.find('div', class_='wprm-recipe-ingredient-group').get_text())
    except:
        ingredients.append(None)
    try:
        instructions.append(soup.find('div', class_='wprm-internal-container wprm-internal-container-instructions').get_text())
    except:
        instructions.append(None)
    try:
        nutrition.append(soup.find('div', class_='wprm-nutrition-label-shortcode-container').get_text())
    except:
        nutrition.append(None)

    print(count)
    print(a.find('h2', class_='wprm-recipe-name wprm-block-text-bold').text)
    count += 1
```

```python
df = pd.DataFrame({
    'name': name,
    'image' : image,
    'description': description,
    'prep_time': prep_time,
    'cook_time': cook_time,
    'total_time': total_time,
    'servings_number': servings_number,
    'servings_unit': servings_unit,
    'course': course,
    'cuisine': cuisine,
    'diet': diet,
    'ingredients': ingredients,
    'instructions': instructions,
    'nutrition': nutrition
})
```

```python
print("name:", len(name))
print("description:", len(description))
print("prep_time:", len(prep_time))
print("cook_time:", len(cook_time))
print("total_time:", len(total_time))
print("servings_number:", len(servings_number))
print("servings_unit:", len(servings_unit))
print("course:", len(course))
print("cuisine:", len(cuisine))
print("diet:", len(diet))
print("ingredients:", len(ingredients))
print("instructions:", len(instructions))
print("nutrition:", len(nutrition))
```

```python
df.to_csv('south_indian_dishes_information.csv')
```
