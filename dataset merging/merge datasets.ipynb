```python
import pandas as pd
import glob

# Get all CSV files ending with _details.csv in the current folder
files = glob.glob("*_details.csv")

# Read and combine
df_list = [pd.read_csv(f) for f in files]
final_df = pd.concat(df_list, ignore_index=True)

# Save merged dataset
final_df.to_csv("merged_details.csv", index=False)

final_df.head()

```
